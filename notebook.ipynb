```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Load the dataset
xmas_movies = pd.read_csv('data/christmas_movies.csv')

# Process the genre data
genre_list = xmas_movies['genre'].str.split(', ').explode()
genre_counts = genre_list.value_counts()

# Plot genre counts
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts, y=genre_counts.index)
plt.title('Frequency of Genres in Christmas Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.show()
```
